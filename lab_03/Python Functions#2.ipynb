```python
def is_above_5_5(movie):
    return movie["imdb"] > 5.5


def filter_movies_above_5_5(movies):
    return [movie for movie in movies if movie["imdb"] > 5.5]  # Fix: 'movie' -> 'movies'


def filter_movies_by_category(movies, category):
    return [movie for movie in movies if movie["category"].lower() == category.lower()]


def average_imdb_score(movies):
    total_score = sum(movie["imdb"] for movie in movies)
    return total_score / len(movies) if movies else 0


def average_imdb_score_by_category(movies, category):
    category_movies = filter_movies_by_category(movies, category)
    return average_imdb_score(category_movies)


print(is_above_5_5(movies[8])) 
print(filter_movies_above_5_5(movies))  
print(filter_movies_by_category(movies, "Romance"))  
print(average_imdb_score(movies)) 
print(average_imdb_score_by_category(movies, "Romance")) 
```
